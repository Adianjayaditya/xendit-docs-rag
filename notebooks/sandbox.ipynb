```python
import os
import yaml

BASE_DIR = r"C:\rag-midtrans-docs\data\raw\docs"

def clean_internal_links(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read()

    # split frontmatter
    if not content.startswith("---"):
        return  # skip kalau tidak ada metadata

    parts = content.split("---", 2)

    if len(parts) < 3:
        return

    _, frontmatter, body = parts

    try:
        metadata = yaml.safe_load(frontmatter)

        if not isinstance(metadata, dict):
            return

        # 🔥 HAPUS internal_links
        if "internal_links" in metadata:
            del metadata["internal_links"]

        # rebuild YAML
        new_frontmatter = yaml.dump(metadata, allow_unicode=True, sort_keys=False)

        new_content = f"---\n{new_frontmatter}---\n{body.lstrip()}"

        with open(file_path, "w", encoding="utf-8") as f:
            f.write(new_content)

        print(f"✅ Cleaned: {os.path.basename(file_path)}")

    except Exception as e:
        print(f"❌ Error di {file_path}: {e}")


def process_all_files():
    for filename in os.listdir(BASE_DIR):
        if filename.endswith(".md"):
            file_path = os.path.join(BASE_DIR, filename)
            clean_internal_links(file_path)


if __name__ == "__main__":
    process_all_files()
```
